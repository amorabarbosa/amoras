```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tabela= pd.read_csv ("/content/Dados oferta 2024.2 - Disciplinas A-Z.csv")
print (tabela)
```

```python
tabela.head(10)
```

```python
# @title Período

from matplotlib import pyplot as plt
tabela['Período\nMatriz'].plot(kind='hist', bins=20, title='Período\nMatriz')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
tabela_sorted = tabela.sort_values(by='Início')
print(tabela_sorted)
```

```python
disciplinas_por_data = tabela_sorted['Início'].value_counts()
plt.figure(figsize=(10, 6))
disciplinas_por_data.plot(kind='bar')
plt.title('Número de Disciplinas por Data de Início')
plt.xlabel('Início')
plt.ylabel('Disciplinas')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
for i in range(len(disciplinas_por_data)):
    plt.text(x=i, y=disciplinas_por_data[i] + 0.1, s=str(disciplinas_por_data[i]), ha='center')
plt.grid(axis='y')  # Adiciona linhas de grade apenas no eixo y
plt.tight_layout()  # Ajusta o layout para evitar cortar as legendas
plt.show()
```

```python
# Remover linhas duplicadas com base nas colunas de disciplina
tabela_unique_disciplinas = tabela_sorted.drop_duplicates(subset=['Disciplinas'])

# Contar o número de disciplinas únicas
num_disciplinas = len(tabela_unique_disciplinas)

print("Número de disciplinas excluindo duplicatas:", num_disciplinas)
```
